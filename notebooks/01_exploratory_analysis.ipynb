# 01. Exploratory Analysis
Анализ бенчмарков и сравнение мономиальных порядков.

```python
import sys; sys.path.insert(0, "..")
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from feature_extractor import BENCHMARK_SYSTEMS, PolynomialSystemFeatureExtractor
from rl_monomial_agent import RLEnvironment
sns.set_style("whitegrid")
```

## 1. Данные бенчмарков

```python
bench = RLEnvironment.BENCHMARK_TIMES
rows = []
for sys_name, times in bench.items():
    for order, t in times.items():
        rows.append({"system": sys_name, "order": order.upper(), "time": t,
                     "speedup_vs_lex": times["lex"] / t})
df = pd.DataFrame(rows)
print(df.to_string(index=False))
```

## 2. Время вычисления

```python
pivot = df.pivot(index="system", columns="order", values="time")
pivot.plot(kind="bar", figsize=(12,5), color=["#3498db","#e74c3c","#2ecc71"])
plt.title("Время вычисления базиса Грёбнера")
plt.ylabel("Время (сек)"); plt.xlabel("Система")
plt.xticks(rotation=45); plt.tight_layout(); plt.show()
```

## 3. Тепловая карта признаков

```python
ext = PolynomialSystemFeatureExtractor()
feat_matrix = np.stack([ext.extract(system_dict=v) for v in BENCHMARK_SYSTEMS.values()])
fig, ax = plt.subplots(figsize=(10, 5))
im = ax.imshow(feat_matrix, aspect="auto", cmap="YlOrRd")
ax.set_xticks(range(8)); ax.set_xticklabels(ext.feature_names(), rotation=45, ha="right")
ax.set_yticks(range(len(BENCHMARK_SYSTEMS)))
ax.set_yticklabels(list(BENCHMARK_SYSTEMS.keys()))
plt.colorbar(im, ax=ax)
plt.title("Признаки полиномиальных систем"); plt.tight_layout(); plt.show()
```
